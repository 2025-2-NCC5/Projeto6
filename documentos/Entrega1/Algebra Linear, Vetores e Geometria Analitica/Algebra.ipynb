```python
import pandas as pd
import numpy as np
```

```python
url_cannoli = "https://github.com/2025-2-NCC5/Projeto6/raw/main/documentos/dados_cannoli.xlsx"

df_cannoli = pd.read_excel(url_cannoli, engine='openpyxl')

display(df_cannoli.head())
display(df_cannoli.info())
```

```python
numerical_columns = ['preparationTime', 'takeOutTimeInSeconds', 'totalAmount']
df_numerical_subset = df_cannoli[numerical_columns]

display(df_numerical_subset.head())
```

Converte os dados numéricos selecionados em matrizes NumPy (vetores e matrizes) e imprima suas formas.

```python
data_matrix = df_numerical_subset.values

data_vector = data_matrix[0, :]

print("Shape of data_matrix:", data_matrix.shape)
print("Shape of data_vector:", data_vector.shape)
```

Implementar e demonstrar operações básicas de álgebra linear (por exemplo, adição, multiplicação, transposição) usando NumPy nos vetores e matrizes criados.

```python
# Certifica-se de que todas as colunas no subconjunto numérico são numéricas
for col in df_numerical_subset.columns:
    df_numerical_subset[col] = pd.to_numeric(df_numerical_subset[col], errors='coerce')

# Remove linhas com valores NaN que podem ter resultado da coerção
df_numerical_subset = df_numerical_subset.dropna()

# Converte o DataFrame numérico limpo em um array NumPy (matriz)
data_matrix = df_numerical_subset.values

# 1. Demonstra a transposição da matriz
data_matrix_transposed = data_matrix.T
print("Forma da matriz transposta:", data_matrix_transposed.shape)

# 2. Seleciona duas linhas para representar vetores
# Garante que há pelo menos duas linhas após remover NaNs
if data_matrix.shape[0] >= 2:
    vector1 = data_matrix[0, :]
    vector2 = data_matrix[1, :]

    # 3. Realiza a adição de vetores
    vector_sum = vector1 + vector2
    print("\nResultado da adição de vetores:")
    print(vector_sum)
else:
    print("\nNão há linhas suficientes na data_matrix para realizar a adição de vetores.")


# 4. Seleciona uma coluna para representar um vetor
# Garante que há pelo menos uma coluna
if data_matrix.shape[1] >= 1:
    column_vector = data_matrix[:, 0]

    # 5. Realiza a multiplicação escalar
    scalar = 2
    scaled_vector = column_vector * scalar
    print("\nResultado da multiplicação escalar:")
    print(scaled_vector)
else:
    print("\nNão há colunas suficientes na data_matrix para selecionar um vetor coluna.")

# 6. Seleciona duas matrizes compatíveis e realiza a multiplicação de matrizes
# Seleciona as 2 primeiras linhas e as 2 primeiras colunas
if data_matrix.shape[0] >= 2 and data_matrix.shape[1] >= 2:
    matrix1 = data_matrix[:2, :2]
    # Seleciona as 2 primeiras linhas e as 2 últimas colunas
    matrix2 = data_matrix[:2, 1:]

    # Realiza a multiplicação de matrizes
    # Verifica se as matrizes são compatíveis para multiplicação (o número de colunas na matrix1 deve ser igual ao número de linhas na matrix2)
    if matrix1.shape[1] == matrix2.shape[0]:
        matrix_product = np.dot(matrix1, matrix2)
        print("\nResultado da multiplicação de matrizes:")
        print(matrix_product)
        print("Forma da matriz resultante:", matrix_product.shape)
    else:
        print("\nAs matrizes não são compatíveis para multiplicação.")
        print("Forma da matrix1:", matrix1.shape)
        print("Forma da matrix2:", matrix2.shape)
else:
    print("\nA matriz de dados é muito pequena para selecionar matrizes compatíveis para multiplicação.")
```
